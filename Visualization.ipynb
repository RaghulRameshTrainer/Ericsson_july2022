# Bar Graph

# Stack Plot

```python
days=[1,2,3,4,5]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]

plt.plot([],[],color='m',label='Working',linewidth=5)
plt.plot([],[],color='b',label='Eating',linewidth=5)
plt.plot([],[],color='r',label='Playing',linewidth=5)
plt.plot([],[],color='k',label='Sleeping',linewidth=5)

plt.stackplot(days,working,eating,playing,sleeping,colors=['m','b','r','k'])
plt.legend()
plt.show()
```

# PIE CHART

```python
slices=[7,2,2,9,4]
activity=['Sleeping','Eating','Playing','Working','Movies']
cols=['c','m','r','b','g']

plt.pie(slices,
        labels=activity,
        colors=cols,
        startangle=90,
        autopct='%1.1f%%',
        explode=(0,0,0.1,0,0)
       )
plt.plot()
```

```python
import numpy as np

def func(x):
    return np.exp(-x)*np.cos(2*np.pi*x)

d1=np.arange(0.0,5.0,0.1)
d2=np.arange(0.0,5.0,0.02)

plt.subplot(211)    #(horizontal, vertical, diagram_position)
plt.plot(d1,func(d1),'bo')

plt.subplot(212)
plt.pie(slices,
        labels=activity,
        colors=cols,
        startangle=90,
        autopct='%1.1f%%',
        explode=(0,0,0.1,0,0)
       )
plt.plot(d1,np.cos(2*np.pi*d1))

plt.show()
```
